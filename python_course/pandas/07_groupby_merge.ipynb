```python
import pandas as pd

data = {
    'Concept': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B'],
    'Value': [10, 20, 30, 40, 50, 60, 70, 80],
    'Category': ['X', 'Y', 'X', 'Z', 'Y', 'Z', 'X', 'Y']
}

df = pd.DataFrame(data)
print(df)
```

```python
grouped = df.groupby('Concept').sum()
print(grouped)# Grouping by a Single Column
# To group by a single column and calculate the sum of each group:


```

```python
# Grouping by Multiple Columns
# You can also group by multiple columns. For example, to group by 'Concept' and 'Category':

grouped = df.groupby(['Concept', 'Category']).sum()
print(grouped)
```

```python
# Applying Different Aggregations
# You can apply different aggregation functions using the agg method:

df.groupby("Concept").agg({
    "Value" : ["mean","median","sum"]   # in his code we not apply ["mean","median","sum"] on "Category" column
})
```

```python
df.groupby("Concept").max()
```

```python
try :
    df.groupby("concept").mean()  
except Exception as e :
    print("error ")

# this error show becoz the mean function also applied on "Categort" column which is object
```

```python
df.groupby("Concept")["Value"].mean()  # that means apply mean() only on value column
```

```python
# select only numeric columns 
numeric_columns = df.select_dtypes(include=["number"]).columns
numeric_columns

# only one column " Value" has numeris values
```

```python
# now you can apply mathematical functions on dataframe using numeric_columns

grouped = df.groupby("Concept")

grouped[numeric_columns].mean()
```

```python
data1 = pd.DataFrame( {
    'key1':[1,2,4,5,6],
    'key2':[4,5,6,7,8],
    'key3':[3,4,5,6,6]

})

data2 = pd.DataFrame( {'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]

})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2)   # both have key1 column , so only same values of key1 columns are showing


'''The pd.merge() function in pandas is used to merge two DataFrames 
based on a common column or index. '''
```

```python
pd.merge(data1,data2,how="left")  

'''how="left": Specifies a left join. In a left join, all rows from data1 (the left DataFrame) 
are included in the result, and matched rows from data2 (the right DataFrame) are added. 
If there is no match, NaN values are used for columns from data2.'''
```

```python
pd.merge(data1,data2,how="right")  # in this key 1 is a data2 table column
```

```python

'''how='outer': Specifies an outer join. An outer join returns all rows from both DataFrames, 
with matching rows from both sides where available. When there is no match, the result is 
NaN for the columns from the DataFrame without a matching row.

on='key1': Specifies the column to join on. If this parameter is omitted, pandas attempts 
to join on columns with the same name in both DataFrames.'''


pd.merge(data1,data2,how= 'outer',on= 'key1')
```

```python
pd.merge(data1,data2,how = 'cross')
```

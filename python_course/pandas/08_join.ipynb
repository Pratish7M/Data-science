```python
import pandas as pd

data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]},
                     index = ['a','b','c','d','e']
)

data2 = pd.DataFrame({'key11':[1,2,4,5,6],
                      'key22':[4,5,6,7,8],
                      'key33':[3,4,5,6,6]
},index=['a','b','h','i','j']
)
```

```python
data1
```

```python
data2
```

```python
data1.join(data2)
```

```python
data1.join(data2,how="right")
```

```python
data2.join(data1, how="inner")
```

```python
data2.join(data1, how="outer")
```
